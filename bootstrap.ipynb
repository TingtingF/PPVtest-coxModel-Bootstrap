```r
Conf(x = df$Norscir, ref = df$Gullstandard,boot = TRUE,
  boot_samples = 1000L)
```

```r
sample_ppv = function(data, indices){
 sample = data[indices, ]
 bar = ml_test(sample$Norscir, sample$Gullstandard, output.as.table = TRUE)$.specificity
 return(bar)
}
```

```r
boot(df, sample_ppv, 500)
```
