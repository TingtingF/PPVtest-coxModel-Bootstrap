**Steg 5: bruke "ml_test" for beregning av PPV**
****Package ‘mltest’
Description: A fast, robust and easy-to-use calculation of multiclass classification evaluation metrics based on confusion matrix****
**https://cran.r-project.org/web/packages/mltest/mltest.pdf**

```r
ml_test(df$Norscir, df$Gullstandard, output.as.table = FALSE)
```

**Steg 6: a bedre fremstilling av PPV via "Conf (Conf: Confusion Matrix And Associated Statistics)"**

```r

Conf(x = df$Norscir, ref = df$Gullstandard)
```

# **Bare "Accuracy" har 95%CI. Men hensikten vår er å beregne PPV. Accuracy = TP+TN/TP+FP+FN+TN. PPV = TP/(TP+FP).
# Spørsmål: Hvordan kan vi ha 95% CI for PPV?? BOOTSTRAPING?**

**Et alternativ å beregner PPV med CI via binom.test**

```r
 binom.test(4,7)# For variabel 1.
```

**Samlet 95% CI for den variabelen**
**(4+5+6+3+5)/35**

```r
binom.test (23,35)
```
